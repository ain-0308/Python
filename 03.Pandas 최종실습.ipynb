```python
# 1.  
import pandas as pd
```

```python
# 2.  
#  data2019, data2020, data2021
data2019 = pd.read_csv('./data/2019.csv', index_col='', encoding='euc-kr')
data2020 = pd.read_csv('./data/2020.csv', index_col='', encoding='euc-kr')
data2021 = pd.read_csv('./data/2021.csv', index_col='', encoding='euc-kr')

#  ''   -  encoding='euc-kr'
```

```python
# 3.      ( 2021 ->  )
data2021.drop('', inplace=True)
data2021
```

```python
# 4.        ''  
data19 = data2019.iloc[:,1:-1]
data2019[''] = data19.sum(axis=1)
# data2019

data20 = data2020.iloc[:,1:6]
data2020[''] = data20.sum(axis=1)


data21 = data2021.iloc[:,1:6]
data2021[''] = data21.sum(axis=1)
data2021
```

```python
# 5. ''   ''     
data2019_1 = data2019[data2019[''] == '']
data2020_1 = data2020[data2020[''] == '']
data2021_1 = data2021[data2021[''] == '']

data2019_1
```

```python
# 6.      - Serise  .
data2019_1['']
data2020_1['']
data2021_1['']
```

```python
# 7.     ( 2019, 2020, 2021)
# data2019.rename(columns={'':'2019'}, inplace=True)
# # data2019
# data2020.rename(columns={'':'2020'}, inplace=True)
# # data2020
# data2021.rename(columns={'':'2021'}, inplace=True)
# data2021
data2019_1.rename(columns={'':'2019'}, inplace=True)
# data2019
data2020_1.rename(columns={'':'2020'}, inplace=True)
# data2020
data2021_1.rename(columns={'':'2021'}, inplace=True)
data2021_1
```

```python
# 8.  
(data2020_1['2020'] - data2019_1['2019'])/data2019_1['2019']*100
(data2021_1['2021'] - data2020_1['2020'])/data2020_1['2020']*100
```

```python
# 9.  DataFrame   ( )
df2019_1 = data2019_1['2019']
df19_20 = (data2020_1['2020'] - data2019_1['2019'])/data2019_1['2019']*100
df2020_1 = data2020_1['2020']
df20_21 = (data2021_1['2021'] - data2020_1['2020'])/data2020_1['2020']*100
df2021_1 = data2021_1['2021']


df = pd.concat([df2019_1, df19_20, df2020_1, df20_21, df2021_1], axis=1)
#rename(columns={'':'2019'}, inplace=True)
df.rename(columns={0:'2019~2020 '}, inplace=True)
df.rename(columns={1:'2020~2021 '}, inplace=True)
df
```
