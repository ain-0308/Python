### 최종 실습
- 범죄 데이터를 활용하여 각 연도별 범죄 증감률을 구해보자

- 증감률 구하는 공식


```python
# 1. 라이브러리 불러오기
import pandas as pd
```

```python
# 2. 데이터 불러오기
# 변수명 data2019, data2020, data2021
data2019 = pd.read_csv('./data/2019.csv', index_col='관서명', encoding='euc-kr')
data2020 = pd.read_csv('./data/2020.csv', index_col='관서명', encoding='euc-kr')
data2021 = pd.read_csv('./data/2021.csv', index_col='관서명', encoding='euc-kr')

# 컬럼중 '관서명' 인덱스 설정 - 인코딩 encoding='euc-kr'
```

```python
# 3. 한개의 년도에만 존재하는 경찰청 삭제 ( 2021년도 -> 광주지방경찰청 )
data2021.drop('광주지방경찰청', inplace=True)
data2021
```

```python
# 4. 각 연도별로 범죄 발생 수의 합계를 구하여 '총합' 컬럼 추가하기
data19 = data2019.iloc[:,1:-1]
data2019['총합'] = data19.sum(axis=1)
# data2019

data20 = data2020.iloc[:,1:6]
data2020['총합'] = data20.sum(axis=1)


data21 = data2021.iloc[:,1:6]
data2021['총합'] = data21.sum(axis=1)
data2021
```

```python
# 5. '구분' 컬럼의 값이 '발생건수' 인 데이터만 출력하여 변수에 담기
data2019_1 = data2019[data2019['구분'] == '발생건수']
data2020_1 = data2020[data2020['구분'] == '발생건수']
data2021_1 = data2021[data2021['구분'] == '발생건수']

data2019_1
```

```python
# 6. 각 연도별로 총합 데이터만 추출 - Serise 형태로 출력될것입니다.
data2019_1['총합']
data2020_1['총합']
data2021_1['총합']
```

```python
# 7. 총합 데이터 이름 변경해주기 ( 2019총계, 2020총계, 2021총계)
# data2019.rename(columns={'총합':'2019총계'}, inplace=True)
# # data2019
# data2020.rename(columns={'총합':'2020총계'}, inplace=True)
# # data2020
# data2021.rename(columns={'총합':'2021총계'}, inplace=True)
# data2021
data2019_1.rename(columns={'총합':'2019총계'}, inplace=True)
# data2019
data2020_1.rename(columns={'총합':'2020총계'}, inplace=True)
# data2020
data2021_1.rename(columns={'총합':'2021총계'}, inplace=True)
data2021_1
```

```python
# 8. 증감율 구하기
(data2020_1['2020총계'] - data2019_1['2019총계'])/data2019_1['2019총계']*100
(data2021_1['2021총계'] - data2020_1['2020총계'])/data2020_1['2020총계']*100
```

```python
# 9. 병합하여 DataFrame 으로 만들어주기 (순서 신경쓸것)
df2019_1 = data2019_1['2019총계']
df19_20 = (data2020_1['2020총계'] - data2019_1['2019총계'])/data2019_1['2019총계']*100
df2020_1 = data2020_1['2020총계']
df20_21 = (data2021_1['2021총계'] - data2020_1['2020총계'])/data2020_1['2020총계']*100
df2021_1 = data2021_1['2021총계']


df = pd.concat([df2019_1, df19_20, df2020_1, df20_21, df2021_1], axis=1)
#rename(columns={'총합':'2019총계'}, inplace=True)
df.rename(columns={0:'2019~2020 증감율'}, inplace=True)
df.rename(columns={1:'2020~2021 증감율'}, inplace=True)
df
```
